```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data1 = pd.read_csv('data-1.csv')
data2 = pd.read_csv('data-2.csv')
data3 = pd.read_csv('data-3.csv')
data4 = pd.read_csv('data-4.csv')
data5 = pd.read_csv('data-5.csv')
data6 = pd.read_csv('data-6.csv')
```

```python
print('data1: ')
S_data1 = data1.describe()
# print(data1)
S_data1 = S_data1.loc[['mean', 'std', 'min', 'max'], ['x', 'y']]
print(S_data1)
print('r value between x and y: ', stats.linregress(data1['x'], data1['y']).rvalue.round(10))
plt.xlabel('x')
plt.ylabel('y')
fit = stats.linregress(data1['x'],data1['y'])
data1['fit'] = data1['x']*fit.slope + fit.intercept
plt.plot(data1['x'], data1['y'], 'b.')
plt.plot(data1['x'], data1['fit'], 'r-')
plt.show()
print('Base on the r value and the graph, there is a strong correlation between x and y in data1, and the data points are scattered around a straight line.')
```
